```python
# Aufgabe 2a

import matplotlib.pyplot as pl
import numpy as np

def poly1(x):
    return x**7-14*x**6+84*x**5-280*x**4+560*x**3-672*x**2+448*x-128
    
    
def poly2(x):
    return (x-2)**7

x = np.linspace(1.99, 2.01, 501) # np.linspace da äquidistante Abstände erwartet werden
y1 = np.array([poly1(xi) for xi in x])
y2 = np.array([poly2(xi) for xi in x])

pl.subplot(121)
pl.plot(x, y1)
pl.title("Poly 1")
pl.xlabel('x axis')
pl.ylabel('y axis')
pl.subplot(122)
pl.plot(x, y2)
pl.title("Poly 2")
pl.xlabel('x axis')
pl.ylabel('y axis')
pl.show()

    
```

```python
# Aufgabe 2b
import matplotlib.pyplot as pl
import numpy as np
# Warning bei x == 0 sowie allen werten die als 0 interpretiert werden
def g(x): 
    if x == 0: return 0
    return x * 1/(np.sin(1+x) - np.sin(1))



x = np.arange(-1e-14, 1e-14, 1e-17)
y = np.array([g(xi) for xi in x])

pl.plot(x, y)
pl.title("g(x)")
pl.show()
```

```python
# Aufgabe 2c
import matplotlib.pyplot as pl
import numpy as np

def gNew(x):
    return x/(2*np.cos((2+x)/2)*np.sin(x/2))



x = np.arange(-1e-14, 1e-14, 1e-17)
y = np.array([gNew(xi) for xi in x])

pl.plot(x, y)
pl.title("gNew(x)")
pl.show()
```
